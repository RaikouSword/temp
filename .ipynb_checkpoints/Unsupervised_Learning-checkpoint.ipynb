```python
#Dataset for Clustering Demonstration
import numpy as np
import matplotlib.pyplot as mat
x= np.random.randint(1,100,6)
y = np.random.randint(1,100,6)
print("X: ",x,"Y: ",y)
xy2 = np.array([[x[i],y[i]] for i in range(6)])
print("XY: ",xy2)

#Visualization before clustering
mat.scatter(xy2[:,0],xy2[:,1])
mat.xlabel('X')
mat.ylabel('Y')
mat.show()
```

```python
# K-Means Clustering
from sklearn.cluster import KMeans

kmeans = KMeans(n_clusters=2 ,random_state=54)
labels = kmeans.fit_predict(xy2)

mat.scatter(xy2[:,0],xy2[:,1],c=labels)
mat.scatter(kmeans.cluster_centers_[:,0],kmeans.cluster_centers_[:,1],marker='x')
mat.title('K-Means Clustering')
mat.xlabel('Feature 1')
mat.ylabel('Feature 2')
mat.grid(True)
mat.show()
```

```python
# Elbow Method for optimal K
"""
Elbow method:
    plot k vs wcss
    elbow point gives optimal k
"""

X = np.array([
    [1,2],[2,3],[3,4],[4,5],[6,7],[51,30],[15,17],[13,24]
])

wcss = []
for k in range(1,9):
    km = KMeans(n_clusters=k,random_state=54)
    km.fit(X)
    wcss.append(km.inertia_)

mat.plot(range(1,9),wcss,marker='o')
mat.xlabel("No of clusters (K)")
mat.ylabel("WCSS")
mat.title("Elbow method")
mat.grid(True)
mat.show()
```

```python
# K-Medoids Clustering (not centroids its medoids)

from sklearn_extra.cluster import KMedoids

kmedoids = KMedoids(n_clusters=2,random_state=54)
kmedoids.fit(X)

print("Medoid indices: ",kmedoids.medoid_indices_)
print("Medoids: ")
print(kmedoids.cluster_centers_)

print("\nCluster labels: ")
print(kmedoids.labels_)

# labels = kmedoids.fit_predict(X)

mat.scatter(X[:,0],X[:,1],label='Data points')
mat.scatter(kmedoids.cluster_centers_[:,0],kmedoids.cluster_centers_[:,1],marker='x')
mat.title('K-Means Clustering')
mat.xlabel('Feature 1')
mat.ylabel('Feature 2')
mat.grid(True)
mat.show()
```
