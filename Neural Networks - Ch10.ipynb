```python
def biological_neuron(signals, threshold=1.0):
    soma = sum(signals)
    if soma>=threshold:
        return 1
    return 0
signals  = [0.3,0.4,0.5]
print("Neuron Output: ",biological_neuron(signals))
```

```python
#Artificial neuron with wts+activation
def artificial_neuron(inputs,weights,bias=0):
    y_sum = np.dot(inputs,weights)+bias
    return y_sum
x = np.array([1,0,1])
w = np.array([0.5,-0.6,0.2])
print("Summation output: ", artificial_neuron(x,w,bias=0.1))
```

```python
import math
#Identity fun
def identity(x):
    return x
#Step/Threshold fun
def step(x):
    return 1 if x>0 else 0
#ReLU fun
def threshold(x,theta=0.5):
    return 1 if x>=theta else 0
#ReLU fun
def relu(x):
    return max(0,x)
#Sigmoid fun
def sigmoid(x,k=1):
    return 1/(1+math.exp(-k*x))
#Bipolar Sigmoid
def bipolar_sigmoid(x):
    return (2/(1+math.exp(-x)))
#Hyperbloic Tangent
def tanh_n(x):
    return math.tanh(x)


values = [-2,-1,0,0.5,1,2]
print("Input values: ",values)
print("\n----Identical function----")
for v in values:
    print(v,"->",identity(v))
print("\n----Step/Threshold function----")
for v in values:
    print(v,"->",step(v))
print("\n----ReLU function----")
for v in values:
    print(v,"->",threshold(v))
print("\n----ReLU function----")
for v in values:
    print(v,"->",relu(v))
print("\n----Sigmoid function----")
for v in values:
    print(v,"->",sigmoid(v))
print("\n----Bipolar function----")
for v in values:
    print(v,"->",bipolar_sigmoid(v))
print("\n----Hyperbloic Tangent function----")
for v in values:
    print(v,"->",tanh_n(v))
```

### MeCulloch-Pitts Neuron(OR Logic Example)

```python
def mcp_neuron(x1,x2,w1=1,w2=1,threshold=1):
    y_sum = x1*w1+x2*w2
    return 1 if y_sum>=threshold else 0
#OR gate truth table
for x1 in [0,1]:
    for x2 in [0,1]:
        print(x1,x2,"->",mcp_neuron(x1,x2))
```

### Rosenblatt Perceptron (Binary Classifier)

```python
import numpy as np

class Perceptron:
    def __init__(self,lr=0.1,epochs=10):
        self.lr=lr
        self.epochs=epochs
    def activation(self,x):
        return 1 if x>=0 else -1
    def fit(self,X,y):
        self.weights = np.zeros(X.shape[1])
        self.bias=0
        
        for _ in range(self.epochs):
            for i in range(len(X)):
                linear = np.dot(X[i],self.weights)+self.bias
                y_pred = self.activation(linear)
                
                update= self.lr * (y[i]-y_pred)
                self.weights+=update  * X[i]
                self.bias+=update
    def predict(self,X):
        linear = np.dot(X,self.weights)+self.bias
        return self.activation(linear)

if __name__=="__main__":
    #Sample data
    X = np.array([[2,3],[1,1],[2,1],[-1,-2]])
    y = np.array([1,1,-1,-1])
    
    p = Perceptron()
    p.fit(X,y)
    print("Prediction: ", p.predict(np.array([2,2])))
```

### ADALINE(Delta Rule/ LMS)

```python
class Adaline:
    def __init__(self,lr=0.1,epochs=20):
        self.lr=lr
        self.epochs=epochs
    def fit(self,X,y):
        self.weights = np.zeros(X.shape[1])
        self.bias=0
        
        for _ in range(self.epochs):
            for i in range(len(X)):
                y_sum = np.dot(X[i],self.weights)+self.bias
                error = y[i]-y_sum
                
                self.weights+=self.lr* error*X[i]
                self.bias+=self.lr*error
    def predict(self,X):
        y_sum = np.dot(X,self.weights)+self.bias
        return 1 if y_sum>=0 else -1

if __name__=="__main__":
    X = np.array([[1,2],[2,3],[3,4],[-1,-2]])
    y = np.array([1,1,1,-1])
    
    a = Adaline()
    a.fit(X,y)
    print("Prediction: ",a.predict(np.array([2,2])))
```

### Single Layer Feed Forward Network

```python
import numpy as np
from sklearn.linear_model import Perceptron
from sklearn.model_selection import train_test_split
from sklearn.datasets import make_classification
from sklearn.metrics import accuracy_score

x,y = make_classification(n_samples=200,n_features=2,n_redundant=0,n_clusters_per_class=1)
x_tr,x_te,y_tr,y_te = train_test_split(x,y,test_size=0.3)
model= Perceptron(max_iter=1000,eta0=0.1)
model.fit(x_tr,y_tr)
y_pred = model.predict(x_te)
print("Accuracy: ",accuracy_score(y_te,y_pred))
```

### Multi-Layer Feed Forward Network

```python
import numpy as np
from sklearn.neural_network import MLPClassifier
from sklearn.model_selection import train_test_split
from sklearn.datasets import make_classification
from sklearn.metrics import accuracy_score

x,y = make_classification(n_samples=300,n_features=4,n_redundant=0,random_state=54,n_informative=4)
x_tr,x_te,y_tr,y_te = train_test_split(x,y,test_size=0.25)
model= MLPClassifier(hidden_layer_sizes=(10,5),activation="relu",max_iter=500,learning_rate_init=0.01)
model.fit(x_tr,y_tr)
y_pred = model.predict(x_te)
print("Accuracy: ",accuracy_score(y_te,y_pred))
```

### Back Propagation

```python
x =np.array([[0,0],[1,0],[0,1],[1,1]])
y = np.array([[0],[1],[1],[0]])

np.random.seed(1)
weights = np.random.random((2,1))
learning_rate=0.1

def sigmoid(x):
    return 1/(1+np.exp(-x))
def sigmoid_derivative(x):
    return x*(1-x)
for epoch in range(5000):
    output = sigmoid(np.dot(x,weights))
    error = y-output
    adjustments = learning_rate*np.dot(x.T,error*sigmoid_derivative(output))
    weights+=adjustments

print("Final Output: ",output)
```
