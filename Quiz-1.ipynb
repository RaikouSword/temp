```python
#Q1 - 
import numpy as np
import matplotlib.pyplot as mat
x= np.random.randint(1,100,10)
y = np.random.randint(1,100,10)
print("X: ",x,"Y: ",y)
xy = np.array([[x[i],y[i]] for i in range(10)])
print("XY: ",xy)
mat.scatter(xy[:,0],xy[:,1])
mat.title("Dataset")
mat.xlabel("X")
mat.ylabel("Y")
mat.show()
```

```python
#Q2
from sklearn.cluster import KMeans
wcss = []
for k in range(1,6):
    km = KMeans(n_clusters=k,random_state=54)
    km.fit(xy)
    wcss.append(km.inertia_)
mat.plot(range(1,6),wcss,marker='x')
mat.title("Elbow")
mat.xlabel("K")
mat.ylabel("WCSS")
mat.show()
```

```python
#Q3
kmeans = KMeans(n_clusters=3 ,random_state=54)
labels = kmeans.fit_predict(xy)
mat.scatter(xy[:,0],xy[:,1],c=labels)
mat.scatter(kmeans.cluster_centers_[:,0],kmeans.cluster_centers_[:,1],marker='x')
mat.title("KMeans")
mat.xlabel("X")
mat.ylabel("Y")
mat.show()
```

```python
#Q4
def nand(x1,x2,w1=1,w2=1,threshold=1):
    y_sum = x1*w1+x2*w2
    return 1 if y_sum<=threshold else 0
for x1 in [0,1]:
    for x2 in [0,1]:
        print(x1,x2,"->",nand(x1,x2))
```

```python
#Q5
def step(x):
    return 1 if x>0 else 0
values = [-0.5,0,1]
for v in values:
    print(v,"->",step(v))
```
