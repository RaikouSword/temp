```python
#1
import numpy as np
import matplotlib.pyplot as mat
from sklearn.datasets import make_blobs

x,y = make_blobs(n_samples = 720, centers=3,n_features=2,random_state=54)
mat.scatter(x[:,0],x[:,1],c=y,cmap='viridis',edgecolors='k')
mat.xlabel('x1')
mat.ylabel('x2')
mat.show()
```

```python
#2
from sklearn.cluster import KMeans

i_val= []
k_range = range(1,10)

for k in k_range:
    kmeans = KMeans(n_clusters=k,random_state = 54)
    kmeans.fit(x)
    i_val.append(kmeans.inertia_)

mat.plot(k_range,i_val)
mat.show()

for k,inertia in zip(k_range,i_val):
    print("k=",k," inertia=",inertia)
```

```python
#4
transactions = [
    {"milk","bread","butter"},
    {'bread','butter'},
    {'milk','bread'},
    {'milk','butter'},
    {'bread','butter'}
]
tid_list = {}
for tid,items in enumerate(transactions,start =1):
    for item in items:
        if item not in tid_list:
            tid_list[item]=[]
        tid_list[item].append(tid)
print(tid_list)
for item in tid_list:
    print(item,":",end=" ")
    support = len(tid_list[item])/len(transactions)
    print(support)
```
