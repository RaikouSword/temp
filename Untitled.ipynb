```python
transaction = [
    {"A","B"},
    {"A","C"},
    {"B","C"},
    {"A","B"}
]
min_sup = 2

#step 1 build tid lists vertical format
tid={}
for i,t in enumerate(transaction):
    for item in t:
        if item not in tid:
            tid[item]=set()
        tid[item].add(i)
print("ITD Lists: ",tid)

#step 2: intersect manually (example: A and B)
AB=tid["A"] & tid["B"] #core idea of ELCAT
print("Support of (A,B):",len(AB))

#step 3: Check if frequent
if len(AB) >= min_sup:
    print("(A,B) is frequent")
    
#Implementtaion of Bagging

from sklearn.datasets import make_classification
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score
from sklearn.ensemble import BaggingClassifier

x,y = make_classification(n_samples=500,n_features=5,random_state=123)
x_tr,x_te,y_tr,y_te = train_test_split(x,y,test_size=0.3,random_state=123)

bag = BaggingClassifier(
    estimator = DecisionTreeClassifier(),
    n_estimators=20,
    bootstrap=True,
    random_state=123
)
bag.fit(x_tr,y_tr)
print("Bagging accuracy: ", accuracy_score(y_te,bag.predict(x_te)))

#Implementation of boosting using Adaboost

from sklearn.ensemble import AdaBoostClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score

ada = AdaBoostClassifier(
    estimator=DecisionTreeClassifier(max_depth=1),
    n_estimators = 50,
    learning_rate=1.0,
    random_state=123
)
ada.fit(x_tr,y_tr)
print("AdaBoost accuracy: ", accuracy_score(y_te,ada.predict(x_te)))

#Implementation of Gradient Boosting Classifer

from sklearn.ensemble import GradientBoostingClassifier
from sklearn.metrics import accuracy_score

gbm = GradientBoostingClassifier(
    n_estimators=100,
    learning_rate=1.0,
    random_state=123
)
gbm.fit(x_tr,y_tr)
print("GBM accuracy: ",accuracy_score(y_te,gbm.predict(x_te)))


#Regularization algo

from sklearn.datasets import make_regression
from sklearn.linear_model import ElasticNet
from sklearn.metrics import mean_squared_error

#Create simple regression data
x,y = make_regression(n_samples=200,n_features=5,noise=10,random_state=123)
#Split
x_tr,x_te,y_tr,y_te = train_test_split(x,y,test_size=0.3,random_state=123)
#Elastic Net model
model = ElasticNet(alpha=1.0,l1_ratio=0.5)
#alpha = regularization strength
#l1_ratio = balance (0=L2 only, 1=l1 only)

model.fit(x_tr,y_tr)
#Predict 
y_pred = model.predict(x_te)

print("MSE: ", mean_squared_error(y_te,y_pred))
```
