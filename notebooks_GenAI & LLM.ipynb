```python
# ========================================================== 
# Iris Flower Classification using Decision Tree 
# ==========================================================

# ----------------------------------------------------------
# LIBRARIES USED 
# ---------------------------------------------------------- 
import os # os = Operating System module # Used for file handling operations.
import joblib # joblib is used to save and load machine learning models. # It stores trained model into a file.
import tkinter as tk # tkinter is Python's built-in GUI library. # Used to create windows, labels, buttons, textboxes etc.
from tkinter import messagebox # messagebox creates popup windows. # Used for success message and error message.
from sklearn.datasets import load_iris # load_iris() loads famous Iris dataset. # Dataset has flower measurements.
from sklearn.model_selection import train_test_split # Used to divide dataset into: # 1. Training data # 2. Testing data 
from sklearn.tree import DecisionTreeClassifier # Imports Decision Tree machine learning algorithm. # Used for classification. 

# ----------------------------------------------------------
# FUNCTION 1 : BUILD MODEL 
# ----------------------------------------------------------
def build_model(): 
    # Function named build_model() # Purpose: Train machine learning model. 
    iris = load_iris() # Loads iris flower dataset.
    
    X = iris.data # X = Input features # Contains 4 flower measurements: # Sepal Length # Sepal Width # Petal Length # Petal Width 
    y = iris.target # y = Output labels # 0 = setosa # 1 = versicolor # 2 = virginica 
    
    X_train, X_test, y_train, y_test = train_test_split( 
        X, y, test_size=0.30, random_state=1 
    ) # Splits dataset: # 70% Training data # 30% Testing data # random_state=1 gives same split every run
    
    model = DecisionTreeClassifier() # Creates Decision Tree model.
    model.fit(X_train, y_train) # Trains model using training data.
    
    joblib.dump(model, "iris_model.pkl") # Saves trained model in file: # iris_model.pkl 

# ----------------------------------------------------------
# FUNCTION 2 : CREATE GUI 
# ---------------------------------------------------------- 
def run_gui(): 
    # Function to create graphical window.
    model = joblib.load("iris_model.pkl") # Loads saved model.
    iris = load_iris() # Loads dataset again. # Used to get flower names.
    
    # ----------------------------------------------------------
    # BUTTON FUNCTION 
    # ---------------------------------------------------------- 
    def classify_flower(): 
        # Runs when button clicked. 
        try: # Try block used to catch invalid input.
            features = [[ 
                float(entry1.get()), 
                float(entry2.get()),
                float(entry3.get()), 
                float(entry4.get()) 
            ]] # Reads values from entry boxes. # Converts text into decimal numbers.
            # Example: # [[5.1, 3.5, 1.4, 0.2]]
            
            prediction = model.predict(features)[0] # Predict flower class. # Returns 0 or 1 or 2 
            flower_name = iris.target_names[prediction] # Converts numeric result into name.
            # Example: # 0 = setosa 
            
            messagebox.showinfo( 
                "Prediction Result", f"The Iris flower is: {flower_name}" 
            ) # Popup window showing prediction.
            
        except ValueError: # If user enters wrong text like abc
            messagebox.showerror( 
                "Input Error", "Please enter valid numeric values." 
            ) # Shows error popup.

    # ----------------------------------------------------------
    # MAIN WINDOW 
    # ---------------------------------------------------------- 
    root = tk.Tk() # Creates main GUI window. 
    root.title("Iris Flower Classification") # Sets window title.
    root.geometry("400x320") # Sets window size: # Width = 400 pixels # Height = 320 pixels
    root.resizable(False, False) # Prevent resizing of window. 
    
    # ----------------------------------------------------------
    # HEADING LABEL 
    # ---------------------------------------------------------- 
    tk.Label( 
        root, text="Iris Flower Prediction", font=("Arial", 14, "bold") 
    ).grid(row=0, column=0, columnspan=2, pady=10) # Creates heading text.
    # row=0 means first row # columnspan=2 means covers 2 columns 
    
    # ----------------------------------------------------------
    # LABELS 
    # ----------------------------------------------------------
    tk.Label(root, text="Sepal Length (cm):").grid(row=1, column=0, padx=10, pady=8)
    tk.Label(root, text="Sepal Width (cm):").grid(row=2, column=0, padx=10, pady=8)
    tk.Label(root, text="Petal Length (cm):").grid(row=3, column=0, padx=10, pady=8)
    tk.Label(root, text="Petal Width (cm):").grid(row=4, column=0, padx=10, pady=8)
    # These lines create text labels.
    
    # ----------------------------------------------------------
    # ENTRY BOXES 
    # ----------------------------------------------------------
    entry1 = tk.Entry(root) # Input box for Sepal Length 
    entry2 = tk.Entry(root) # Input box for Sepal Width 
    entry3 = tk.Entry(root) # Input box for Petal Length
    entry4 = tk.Entry(root) # Input box for Petal Width 
    
    # ----------------------------------------------------------
    # POSITION ENTRY BOXES 
    # ---------------------------------------------------------- 
    entry1.grid(row=1, column=1, padx=10, pady=8) 
    entry2.grid(row=2, column=1, padx=10, pady=8)
    entry3.grid(row=3, column=1, padx=10, pady=8) 
    entry4.grid(row=4, column=1, padx=10, pady=8)
    # Places entry boxes beside labels. 
    
    # ----------------------------------------------------------
    # BUTTON 
    # ---------------------------------------------------------- 
    tk.Button( 
        root, text="Classify Flower", width=18, bg="lightblue", command=classify_flower
    ).grid(row=5, column=0, columnspan=2, pady=20) # Creates button.
    # command=classify_flower means: # when clicked run prediction. 
    
    # ----------------------------------------------------------
    # RUN GUI 
    # ----------------------------------------------------------
    root.mainloop() # Keeps window open. # Waits for user actions.

# ----------------------------------------------------------
# MAIN PROGRAM 
# ---------------------------------------------------------- 
if __name__ == "__main__": # Runs only when file is executed directly.
    if not os.path.exists("iris_model.pkl"): # Checks if model file exists. 
        build_model() # If file missing: # train model and save file.
        
    run_gui() # Opens GUI window.
```
